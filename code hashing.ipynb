```python
#!/usr/bin/env python
# coding: utf-8

# In[9]:

count = 0

class Hash_table() :

  def __init__(self, prime, size) :
    self.size = size
    self.prime = prime
    self.hashtable = [None]*self.size
        
  def hash(self,key):
    # Hash function h(x) = x%31
    global count
    index = key%self.prime
    if self.hashtable[index] == None :
      return index
    else :
      # Implementing linear probing
      while self.hashtable[index] != None :
        index = (index+1)%self.size
        count += 1
      return index

  def insert(self,key) :
    index = self.hash(key)
    self.hashtable[index] = key
       
  def search(self,key) :
    global count
    index = key%self.prime
    if self.hashtable[index] != key :
      while self.hashtable[index] != key and self.hashtable[index] != None :
        index = (index+1)%self.size
        count += 1 
    if self.hashtable[index] == key :
      return index
    else :
      return None
   
  def print_hashtable(self) :
    global count
    print("Hash table is :-\n\nindex \t value")
    for x in range(len(self.hashtable)) :
      print(x,"\t",self.hashtable[x],count)
        

# Initializing hash table of size 35
HT = Hash_table(31, 35)


HT.insert(11119)
HT.insert(88888)
HT.insert(25247)
HT.insert(58726)
HT.insert(35355)
HT.insert(11114)
HT.insert(88883)
HT.insert(25242)
HT.insert(58721)
HT.insert(35359)
HT.insert(11118)
HT.insert(88887)
HT.insert(25246)
HT.insert(58725)
HT.insert(35354)
HT.insert(11113)
HT.insert(88882)
HT.insert(25241)
HT.insert(58729)
HT.insert(35358)
HT.insert(11117)
HT.insert(88886)
HT.insert(25245)
HT.insert(58724)
HT.insert(35353)
HT.insert(11112)
HT.insert(88881)
HT.insert(25249)
HT.insert(58728)
HT.insert(35357)
HT.insert(11116)
HT.insert(12345)




HT.print_hashtable()

index = HT.search(12345)

if index!= None :
    print("\nGiven key is present at index",index)
else :
    print("\nGiven key is not present in the hash table")


# In[ ]:
```
